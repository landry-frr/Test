# Etape 3 : Affichage de variables

```java
String message = "Hello";

String Produit1 = "Stylo";
double prixunitaireproduit1 = 1.20;
int quantite1 = 5;
double totalP1 = (prixunitaireproduit1 * quantite1);


String Produit2 = "Cahier";
double prixunitaireproduit2 = 3.50;
int quantite2 = 2;
double totalP2 = (prixunitaireproduit2 * quantite2);

double totalproduits = (totalP1 + totalP2);


System.out.println("+----------------------+-------+-----+--------+");
System.out.println("| Produit              | PU    | Qté |  Prix  |");
System.out.println("+----------------------+-------+-----+--------+");
System.out.printf("| %-20s | %5.2f | %3d | %6.2f |\n", Produit1, prixunitaireproduit1, quantite1, totalP1);
System.out.printf("| %-20s | %5.2f | %3d | %6.2f |\n", Produit2, prixunitaireproduit2, quantite2, totalP2);
System.out.println("+----------------------+-------+-----+--------+");
System.out.printf("| Total                              | %6.2f |\n", totalproduits);
System.out.println("+------------------------------------+--------+");

```

# Etape 4 : Saisie des informations

```java
String message = "Hello";

String Produit1 = "Stylo";
double prixunitaireproduit1 = 1.20;
int quantite1 = 5;

String Produit2 = "Cahier";
double prixunitaireproduit2 = 3.50;
int quantite2 = 2;



Scanner clavier = new Scanner(System.in);


System.out.println("Quantité de stylos ?");
 quantite1 = clavier.nextInt();


System.out.println("Quantité de cahiers ?");
quantite2 = clavier.nextInt();

double totalP1 = (prixunitaireproduit1 * quantite1);


double totalP2 = (prixunitaireproduit2 * quantite2);

double totalproduits = (totalP1 + totalP2);

System.out.println("+----------------------+-------+-----+--------+");
System.out.println("| Produit              | PU    | Qté |  Prix  |");
System.out.println("+----------------------+-------+-----+--------+");
System.out.printf("| %-20s | %5.2f | %3d | %6.2f |\n", Produit1, prixunitaireproduit1, quantite1, totalP1);
System.out.printf("| %-20s | %5.2f | %3d | %6.2f |\n", Produit2, prixunitaireproduit2, quantite2, totalP2);
System.out.println("+----------------------+-------+-----+--------+");
System.out.printf("| Total                              | %6.2f |\n", totalproduits);
System.out.println("+------------------------------------+--------+");
```
